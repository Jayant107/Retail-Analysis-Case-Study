```python
import numpy as np
import pandas as  pd
import matplotlib.pyplot as plt
import seaborn as sns

mom_growth_rate = pd.read_csv("/content/m-o-m_growth_rate.csv")
segmentaion_total_amount_spent = pd.read_csv("/content/segmentaion_total_amount_spent.csv")
segmentaion_total_product_purchased = pd.read_csv("/content/segmentaion_total_product_purchased.csv")
segmentation_loyality = pd.read_csv("/content/segmentation_loyality.csv")
total_sales_per_month = pd.read_csv("/content/total_sales_per_month.csv")
product_sold_per_category = pd.read_csv("/content/product_sold_per_category.csv")
repeat_patterns = pd.read_csv("/content/repeat_patterns.csv")
mom_growth_rate.head(6)
```

```python
mom_growth_rate[mom_growth_rate.isna() > 0].sum()
mom_growth_rate.fillna(0, inplace=True)
```

```python
plt.figure(figsize=(20, 10))
ax = plt.gca()
ax.set_title("Month-Over-Month Growth rate")
ax.plot(mom_growth_rate["yr_m"].head(20), mom_growth_rate["MoM_Growth_Percent"].head(20), color="orange", marker="o")
ax.yaxis.set_major_formatter(lambda x,_: "{:.2%}".format(x/100))
plt.savefig("./mom.png")
plt.show()
```

```python
plt.figure(figsize=(5, 5))
ax = plt.gca()
ax.set_title("Segmentation of customer on total amount spent")
ax.pie(segmentaion_total_amount_spent["Percentage_Per_Segment"], labels=segmentaion_total_amount_spent["Customer_Segment"], autopct='%1.1f%%')
plt.savefig("seg_of_cus_tsp")
plt.show()
```

```python
plt.figure(figsize=(5, 5))
ax = plt.gca()
ax.set_title("Segmentation of customer on total product purchased")
ax.pie(segmentaion_total_product_purchased["Percentage_Per_Segment"], labels=segmentaion_total_product_purchased["Customer_Segment"], autopct='%1.1f%%')
plt.savefig("seg_of_cus_tpp")
plt.show()
```

```python
plt.figure(figsize=(5, 5))
ax = plt.gca()
ax.set_title("Segmentation of customer on loyalty")
ax.pie(segmentation_loyality["Percentage"], labels=segmentation_loyality["Loyalty_Status"], autopct='%1.1f%%')
plt.savefig("seg_of_cus_loyalty.png")
plt.show()
```
